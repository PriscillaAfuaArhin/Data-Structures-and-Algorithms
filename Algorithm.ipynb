##DAY 1
Given an array of integers,and an integer target,return indices of two numbers such that they add up to get the target.

FUNCTION twoSum(nums, target):

     Step 1: Pick the first number
    FOR i FROM 0 TO length of nums - 1:
    
        Step 2: Pick the second number (start from i + 1)
        FOR j FROM i + 1 TO length of nums - 1:
        
            Step 3: Check if they add up to target
            IF nums[i] + nums[j] EQUALS target:
            
                Step 4: Found the answer, return indices
                RETURN [i, j]
                
            END IF
            
        END FOR
        
    END FOR
    
    Step 5: If no solution is found (optional based on constraints)
    RETURN "No solution found"

END FUNCTION




```python
# Day 1 Code
def twoSum(nums, target):
    # Step 1: Pick the first number using its index
    for i in range(len(nums)):
        # Step 2: Pick the second number, starting from the next index
        for j in range(i + 1, len(nums)):
            # Step 3: Check if they add up to target
            if nums[i] + nums[j] == target:
                # Step 4: Found the answer, return indices
                return [i, j]
    # Step 5: If no solution is found after checking all pairs
    return "No solution found"

target = 9
numbers = [2, 7, 11, 15]

result_indices = twoSum(numbers, target)
print(f"The indices of the numbers that add up to {target} are: {result_indices}")
```

##DAY 2
Given an array of integer numbers,find the sub array with the largest sum and return its sum

FUNCTION maxSubArray(nums):

     Step 1: Set initial values
    current_sum = nums[0]
    max_sum = nums[0]

     Step 2: Loop starting from the 2nd number
    FOR i FROM 1 TO length of nums - 1:
        
        number = nums[i]

         Step 3: Decision - Start fresh or accumulate?
         We take the larger of the two: (number) OR (current_sum + number)
        current_sum = MAX(number, current_sum + number)
        
         Step 4: Record keeping
         If our current chain is the best we've seen, remember it
        IF current_sum > max_sum:
            max_sum = current_sum

    END FOR

     Step 5: Final answer
    RETURN max_sum

END FUNCTION

```python
#Day 2 code
def maxSubArray(nums):
    # Step 1: Set initial values
    current_sum = nums[0]
    max_sum = nums[0]

    # Step 2: Loop starting from the 2nd number
    for i in range(1, len(nums)):
        number = nums[i]

        # Step 3: Decision - Start fresh or accumulate?
        # We take the larger of the two: (number) OR (current_sum + number)
        current_sum = max(number, current_sum + number)

        # Step 4: Record keeping
        # If our current chain is the best we've seen, remember it
        if current_sum > max_sum:
            max_sum = current_sum

    # Step 5: Final answer
    return max_sum

# Example usage:
numbers_day2 = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
result_max_sum = maxSubArray(numbers_day2)
print(f"The maximum subarray sum is: {result_max_sum}")

numbers_day2_2 = [1]
result_max_sum_2 = maxSubArray(numbers_day2_2)
print(f"The maximum subarray sum for [1] is: {result_max_sum_2}")

numbers_day2_3 = [5, 4, -1, 7, 8]
result_max_sum_3 = maxSubArray(numbers_day2_3)
print(f"The maximum subarray sum for [5, 4, -1, 7, 8] is: {result_max_sum_3}")
```

            Found a 2: Send it to the High zone
            SWAP(nums[mid], nums[high])
            high = high - 1
             Note: We do NOT increment mid here
            
    END WHILE

END FUNCTION

FUNCTION fourSum(nums, target):

    // Step 1: Sort the array to handle duplicates and use pointers
    SORT(nums)
    
    // Create a list to store the answers
    results = []
    n = length of nums

    // Step 2: Fix the first number (i)
    FOR i FROM 0 TO n - 1:
        
        // Skip duplicate numbers for the first position
        IF i > 0 AND nums[i] == nums[i-1]:
            CONTINUE to next i

        // Step 3: Fix the second number (j)
        FOR j FROM i + 1 TO n - 1:
            
            // Skip duplicate numbers for the second position
            IF j > i + 1 AND nums[j] == nums[j-1]:
                CONTINUE to next j

            // Step 4: Use Pointers for the 3rd and 4th numbers
            left = j + 1
            right = n - 1

            WHILE left < right:
                
                sum = nums[i] + nums[j] + nums[left] + nums[right]

                IF sum == target:
                    // Found a match! Add to results
                    ADD [nums[i], nums[j], nums[left], nums[right]] TO results

                    // Move pointers inward to find more pairs
                    left = left + 1
                    right = right - 1

                    // Skip duplicates for the 3rd number (Left Pointer)
                    WHILE left < right AND nums[left] == nums[left-1]:
                        left = left + 1

                    // Skip duplicates for the 4th number (Right Pointer)
                    WHILE left < right AND nums[right] == nums[right+1]:
                        right = right - 1

                ELSE IF sum < target:
                    // Sum is too small, need a bigger number
                    left = left + 1

                ELSE:
                    // Sum is too big, need a smaller number
                    right = right - 1

            END WHILE
        
        END FOR
    
    END FOR

    RETURN results

END FUNCTION

```python
#Code
class Solution:
    def fourSum(self, nums: list[int], target: int) -> list[list[int]]:
        # Step 1: Sort the array (Crucial for two pointers)
        nums.sort()
        results = []
        n = len(nums)

        # Step 2: First Loop (Fix the 1st number)
        for i in range(n):
            # Skip duplicates for the 1st number
            if i > 0 and nums[i] == nums[i-1]:
                continue

            # Step 3: Second Loop (Fix the 2nd number)
            # Start from i + 1 to avoid reusing the 1st number
            for j in range(i + 1, n):
                # Skip duplicates for the 2nd number
                if j > i + 1 and nums[j] == nums[j-1]:
                    continue

                # Step 4: Two Pointers (Find the 3rd and 4th numbers)
                left = j + 1
                right = n - 1

                while left < right:
                    total = nums[i] + nums[j] + nums[left] + nums[right]

                    if total == target:
                        # Found a valid quadruplet!
                        results.append([nums[i], nums[j], nums[left], nums[right]])

                        # Move pointers inward
                        left += 1
                        right -= 1

                        # Skip duplicates for the 3rd number (left pointer)
                        while left < right and nums[left] == nums[left-1]:
                            left += 1

                        # Skip duplicates for the 4th number (right pointer)
                        while left < right and nums[right] == nums[right+1]:
                            right -= 1

                    elif total < target:
                        # Sum too small -> Need bigger number -> Move Left forward
                        left += 1
                    else:
                        # Sum too big -> Need smaller number -> Move Right backward
                        right -= 1

        return results

# --- Example usage (moved outside the class) ---
solver = Solution() # Create an instance of the Solution class

nums1 = [1, 0, -1, 0, -2, 2]
target1 = 0
print(f"Example 1 Input: {nums1}, Target: {target1}")
print(f"Result: {solver.fourSum(nums1, target1)}")
print("-" * 30)

nums2 = [2, 2, 2, 2, 2]
target2 = 8
print(f"Example 2 Input: {nums2}, Target: {target2}")
print(f"Result: {solver.fourSum(nums2, target2)}")
print("-" * 30)
```
