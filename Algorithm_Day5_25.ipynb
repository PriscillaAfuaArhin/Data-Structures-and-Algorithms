#DAY 5


1. Sort intervals by start time
2. Create empty list called merged
3. For each interval:
      - If merged is empty, add interval
      - Else if current interval starts after last interval ends:
            add interval
      - Else:
            merge intervals by updating the end time
4. Return merged list


```python
def merge_intervals(intervals):
    # Step 1: Sort intervals based on the starting value
    intervals.sort()

    # Step 2: Create an empty list to store merged intervals
    merged = []

    # Step 3: Loop through each interval
    for interval in intervals:
        # If merged is empty, add the first interval
        if not merged:
            merged.append(interval)

        # Check if current interval overlaps with the last one
        elif interval[0] <= merged[-1][1]:
            # Merge by updating the end value
            merged[-1][1] = max(merged[-1][1], interval[1])

        # If no overlap, add the interval
        else:
            merged.append(interval)

    # Step 4: Return the merged intervals
    return merged


# Example input
intervals = [[1, 3], [2, 6], [8, 10], [15, 18]]

# Function call
result = merge_intervals(intervals)

# Output
print(result)
```

##Day 6

1. Create an empty list result
2. Set open_count = 0

3. Loop through the string (left to right):
     - If character is '(':
           add it to result
           open_count += 1
     - If character is ')':
           if open_count > 0:
               add it to result
               open_count -= 1
           else:
               skip it
     - Else (letter):
           add it to result

4. Create final_result (empty list)
5. Loop through result in reverse order:
     - If character is '(' and open_count > 0:
           skip it
           open_count -= 1
     - Else:
           add it to final_result

6. Reverse final_result and return it as a string


```python
def make_valid_parentheses(s):
    results = []
    open_count = 0

    # First pass: remove extra ')'
    for char in s:
        if char == '(':
            results.append(char)
            open_count += 1
        elif char == ')':
            if open_count > 0:
                results.append(char)
                open_count -= 1
            # else: skip invalid ')'
        else:
            results.append(char)

    final_results = []

    # Second pass: remove extra '(' from the end
    for char in reversed(results):
        if char == '(' and open_count > 0:
            open_count -= 1
            # skip extra '('
        else:
            final_results.append(char)

    # Reverse back and return as string
    return ''.join(reversed(final_results))


# Example usage
s = "a)b(c)d"
print(make_valid_parentheses(s))
```

##Day7

1. Create an empty dictionary to count characters
2. Loop through the string:
      - Increase count of each character
3. Sort the characters by frequency in descending order
4. Create an empty result string
5. For each character in sorted order:
      - Repeat the character according to its frequency
6. Return the result string


```python
def frequency_sort(s):
    # Step 1: Count how many times each character appears
    freq = {}

    for char in s:
        if char in freq:
            freq[char] = freq[char] + 1
        else:
            freq[char] = 1

    # Step 2: Sort characters by frequency (highest first)
    sorted_items = sorted(freq.items(), key=lambda item: item[1], reverse=True)

    # Step 3: Build the final string
    result = ""

    for char, count in sorted_items:
        result = result + (char * count)

    return result


# Example test
text = "tree"
print(frequency_sort(text))
```

##Day 8

1. If s1 is longer than s2:
      return False

2. Count characters in s1 → count_s1

3. For every substring of s2 with length = length of s1:
      - Count characters in the substring
      - If the counts match count_s1:
            return True

4. If no match is found:
      return False


```python
def contains_permutation(s1, s2):
    # If s1 is longer, it cannot be inside s2
    if len(s1) > len(s2):
        return False

    # Step 1: Count characters in s1
    count_s1 = {}
    for char in s1:
        if char in count_s1:
            count_s1[char] += 1
        else:
            count_s1[char] = 1

    # Step 2: Check each substring of s2
    window_length = len(s1)

    for i in range(len(s2) - window_length + 1):
        substring = s2[i:i + window_length]

        # Count characters in the substring
        count_sub = {}
        for char in substring:
            if char in count_sub:
                count_sub[char] += 1
            else:
                count_sub[char] = 1

        # Compare counts
        if count_sub == count_s1:
            return True

    return False


# Example tests
print(contains_permutation("ab", "eidbaooo"))  # True
print(contains_permutation("ab", "eidboaoo"))  # False
```

##Day 9

1. Create an empty list called result

2. Define a helper function:
      - It takes:
            index (current position)
            path (current list of palindromes)

3. If index reaches the end of the string:
      - Add path to result
      - Stop

4. From index to end of string:
      - Take substring
      - If substring is a palindrome:
            add it to path
            call helper again for next index
            remove last substring (backtrack)

5. Start helper from index 0
6. Return result


```python
def palindrome_partition(s):
    result = []

    # Function to check if a string is a palindrome
    def is_palindrome(text):
        return text == text[::-1]

    # Backtracking function
    def backtrack(start, path):
        # If we reach the end of the string
        if start == len(s):
            result.append(path.copy())
            return

        # Try every possible substring
        for end in range(start + 1, len(s) + 1):
            substring = s[start:end]

            # Check if substring is a palindrome
            if is_palindrome(substring):
                path.append(substring)       # choose
                backtrack(end, path)          # explore
                path.pop()                    # un-choose (backtrack)

    # Start from index 0
    backtrack(0, [])

    return result


# Example test
print(palindrome_partition("aab"))
```

##Day 10

1. If t is longer than s:
      return ""

2. Count how many times each character appears in t

3. Set:
      left = 0
      min_length = infinity
      result = ""

4. Loop with right pointer over s:
      - Add s[right] to window count

      - If character count matches what we need:
            increase matched count

5. While all characters are matched:
      - Update result if current window is smaller
      - Remove s[left] from window
      - Move left pointer forward

6. Return result


```python
def min_window(s, t):
    # Edge case
    if len(t) > len(s):
        return ""

    # Step 1: Count characters in t
    need = {}
    for char in t:
        if char in need:
            need[char] += 1
        else:
            need[char] = 1

    # Step 2: Sliding window setup
    window = {}
    left = 0
    matched = 0
    required = len(need)

    min_length = float("inf")
    result = ""

    # Step 3: Expand the window
    for right in range(len(s)):
        char = s[right]
        window[char] = window.get(char, 0) + 1

        # Check if this character is fully matched
        if char in need and window[char] == need[char]:
            matched += 1

        # Step 4: Try to shrink the window
        while matched == required:
            current_length = right - left + 1
            if current_length < min_length:
                min_length = current_length
                result = s[left:right + 1]

            # Remove left character
            left_char = s[left]
            window[left_char] -= 1

            if left_char in need and window[left_char] < need[left_char]:
                matched -= 1

            left += 1

    return result


# Example test
s = "ADOBECODEBANC"
t = "ABC"
print(min_window(s, t))
```

##Day 11

1. If the list is empty:
      return None

2. Recursively process the next node

3. If current node value == val:
      return the cleaned next node (skip current)

4. Else:
      link current node to cleaned next node
      return current node


```python
# Definition for a linked list node
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def remove_elements(head, val):
    # Base case: if list is empty
    if head is None:
        return None

    # Recursively clean the rest of the list
    head.next = remove_elements(head.next, val)

    # Decide whether to keep or remove current node
    if head.val == val:
        return head.next   # skip this node
    else:
        return head        # keep this node
```

##Day 12

1. If the list is empty OR has only one node:
      return head

2. Recursively reverse the rest of the list
      new_head = reverse(head.next)

3. Make the next node point back to current node
      head.next.next = head

4. Set current node's next to None
      head.next = None

5. Return new_head


```python
# Definition for singly-linked list
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def reverse_list(head):
    # Base case: empty list or single node
    if head is None or head.next is None:
        return head

    # Step 1: Reverse the rest of the list
    new_head = reverse_list(head.next)

    # Step 2: Change the direction
    head.next.next = head
    head.next = None

    return new_head
```

##Day 13

1. Create an empty list called result

2. Define a recursive function that takes:
      - index (current position)
      - path (current subset)

3. If index reaches the end of nums:
      - Add a copy of path to result
      - Stop

4. Choice 1: Do NOT include nums[index]
      - Call recursion with next index

5. Choice 2: Include nums[index]
      - Add nums[index] to path
      - Call recursion with next index
      - Remove nums[index] (backtrack)

6. Start recursion from index 0
7. Return result


```python
def subsets(nums):
    result = []

    def backtrack(index, path):
        # If we have considered all elements
        if index == len(nums):
            result.append(path.copy())
            return

        # Choice 1: Do NOT include nums[index]
        backtrack(index + 1, path)

        # Choice 2: Include nums[index]
        path.append(nums[index])
        backtrack(index + 1, path)
        path.pop()  # backtrack

    backtrack(0, [])
    return result


# Example test
nums = [1, 2, 3]
print(subsets(nums))
```

##Day 14

1. Create an empty list result

2. Define a recursive function with:
      - current string
      - number of '(' used
      - number of ')' used

3. If current string length = 2 * n:
      - Add it to result
      - Stop

4. If number of '(' used < n:
      - Add '(' and recurse

5. If number of ')' used < '(' used:
      - Add ')' and recurse

6. Start recursion with empty string
7. Return result


```python
def generate_parentheses(n):
    result = []

    def backtrack(current, open_count, close_count):
        # If the string is complete
        if len(current) == 2 * n:
            result.append(current)
            return

        # Add '(' if we still can
        if open_count < n:
            backtrack(current + "(", open_count + 1, close_count)

        # Add ')' only if it keeps the string valid
        if close_count < open_count:
            backtrack(current + ")", open_count, close_count + 1)

    # Start recursion
    backtrack("", 0, 0)
    return result


# Example test
print(generate_parentheses(3))
```

##Day 15

1. Create a dictionary to store key-value pairs
2. Create a list to track usage order
3. For get(key):
      - If key not in dictionary → return -1
      - Move key to end of list (recently used)
      - Return value
4. For put(key, value):
      - If key exists:
            update value
            move key to end
      - Else:
            add key and value
            add key to list
      - If size exceeds capacity:
            remove first key from list
            delete it from dictionary


```python
class LRUCache:
    def __init__(self, capacity):
        self.capacity = capacity
        self.cache = {}     # stores key-value pairs
        self.order = []     # tracks usage order

    def get(self, key):
        if key not in self.cache:
            return -1

        # Mark key as recently used
        self.order.remove(key)
        self.order.append(key)

        return self.cache[key]

    def put(self, key, value):
        if key in self.cache:
            # Update value and mark as recently used
            self.order.remove(key)
        elif len(self.cache) == self.capacity:
            # Remove least recently used item
            oldest_key = self.order.pop(0)
            del self.cache[oldest_key]

        # Add new key
        self.cache[key] = value
        self.order.append(key)
```

##Day 16

1. Let n = length of nums

2. For each index i from 0 to n-1:
      While nums[i] is between 1 and n
      AND nums[i] is not in the correct position:
            swap nums[i] with nums[nums[i] - 1]

3. After rearranging:
      For each index i from 0 to n-1:
            if nums[i] != i + 1:
                  return i + 1

4. If everything is correct:
      return n + 1


```python
def first_missing_positive(nums):
    n = len(nums)

    # Step 1: Place each number in its correct position
    for i in range(n):
        # Keep swapping until the number is either out of range
        # or already in the correct position
        while 1 <= nums[i] <= n and nums[nums[i] - 1] != nums[i]:
            correct_index = nums[i] - 1
            nums[i], nums[correct_index] = nums[correct_index], nums[i]

    # Step 2: Find the first missing positive
    for i in range(n):
        if nums[i] != i + 1:
            return i + 1

    # Step 3: If all numbers are present
    return n + 1


# Example tests
print(first_missing_positive([1, 2, 0]))
print(first_missing_positive([3, 4, -1, 1]))
print(first_missing_positive([7, 8, 9, 11, 12]))
```

##Day 17

1. If matrix is empty, return empty list

2. Set:
      top = 0
      bottom = number of rows - 1
      left = 0
      right = number of columns - 1

3. While top <= bottom AND left <= right:
      - Traverse top row from left to right
      - Move top down by 1

      - Traverse right column from top to bottom
      - Move right left by 1

      - If top <= bottom:
            traverse bottom row from right to left
            move bottom up by 1

      - If left <= right:
            traverse left column from bottom to top
            move left right by 1

4. Return the result list


```python
def spiral_order(matrix):
    # If the matrix is empty
    if not matrix:
        return []

    result = []

    top = 0
    bottom = len(matrix) - 1
    left = 0
    right = len(matrix[0]) - 1

    # Continue until all elements are visited
    while top <= bottom and left <= right:

        # 1. Traverse from left to right (top row)
        for col in range(left, right + 1):
            result.append(matrix[top][col])
        top += 1

        # 2. Traverse from top to bottom (right column)
        for row in range(top, bottom + 1):
            result.append(matrix[row][right])
        right -= 1

        # 3. Traverse from right to left (bottom row)
        if top <= bottom:
            for col in range(right, left - 1, -1):
                result.append(matrix[bottom][col])
            bottom -= 1

        # 4. Traverse from bottom to top (left column)
        if left <= right:
            for row in range(bottom, top - 1, -1):
                result.append(matrix[row][left])
            left += 1

    return result


# Example test
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(spiral_order(matrix))
```

##Day 18

1. For each row:
      - Create an empty set
      - If a number repeats → return False

2. For each column:
      - Create an empty set
      - If a number repeats → return False

3. For each 3×3 sub-box:
      - Create an empty set
      - If a number repeats → return False

4. If all checks pass:
      return True


```python
def is_valid_sudoku(board):

    # 1. Check rows
    for row in range(9):
        seen = set()
        for col in range(9):
            value = board[row][col]
            if value != ".":
                if value in seen:
                    return False
                seen.add(value)

    # 2. Check columns
    for col in range(9):
        seen = set()
        for row in range(9):
            value = board[row][col]
            if value != ".":
                if value in seen:
                    return False
                seen.add(value)

    # 3. Check 3×3 sub-boxes
    for box_row in range(0, 9, 3):
        for box_col in range(0, 9, 3):
            seen = set()
            for i in range(3):
                for j in range(3):
                    value = board[box_row + i][box_col + j]
                    if value != ".":
                        if value in seen:
                            return False
                        seen.add(value)

    return True
```

##Day 19

1. For every cell in the grid:
      - If it matches the first letter of the word:
            try to find the rest of the word from there

2. To search from a cell:
      - If all letters are matched → return True
      - If out of bounds or letter doesn’t match → return False
      - Mark the cell as visited
      - Try all 4 directions (up, down, left, right)
      - Unmark the cell (backtrack)

3. If no path works:
      return False


```python
def exist(board, word):
    rows = len(board)
    cols = len(board[0])

    def search(row, col, index):
        # If all letters are found
        if index == len(word):
            return True

        # Check boundaries and character match
        if row < 0 or row >= rows or col < 0 or col >= cols:
            return False
        if board[row][col] != word[index]:
            return False

        # Mark the cell as visited
        temp = board[row][col]
        board[row][col] = "#"

        # Explore all directions
        found = (
            search(row + 1, col, index + 1) or
            search(row - 1, col, index + 1) or
            search(row, col + 1, index + 1) or
            search(row, col - 1, index + 1)
        )

        # Restore the cell (backtrack)
        board[row][col] = temp

        return found

    # Try starting from every cell
    for r in range(rows):
        for c in range(cols):
            if search(r, c, 0):
                return True

    return False
```

##Day 20

1. If the tree is empty:
      return

2. Recursively flatten left subtree
3. Recursively flatten right subtree

4. Store the right subtree
5. Move left subtree to the right
6. Set left pointer to None

7. Move to the end of the new right subtree
8. Attach the original right subtree


```python
# Definition for a binary tree node
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def flatten(root):
    # Base case: empty tree
    if root is None:
        return

    # Step 1: Flatten left and right subtrees
    flatten(root.left)
    flatten(root.right)

    # Step 2: Store the original right subtree
    right_subtree = root.right

    # Step 3: Move left subtree to the right
    root.right = root.left
    root.left = None

    # Step 4: Go to the end of the new right subtree
    current = root
    while current.right:
        current = current.right

    # Step 5: Attach the original right subtree
    current.right = right_subtree
```

##Day 21

1. Create an empty list called values
2. Traverse the linked list:
      - Add each node's value to values
3. Check if values is equal to its reverse
4. If yes, return True
5. Otherwise, return False


```python
# Definition for singly-linked list
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def is_palindrome(head):
    values = []
    current = head

    # Step 1: Store values in a list
    while current:
        values.append(current.val)
        current = current.next

    # Step 2: Check if list is a palindrome
    return values == values[::-1]
```

##Day 22

1. Count k nodes from the head
      - If there are fewer than k nodes, return head

2. Reverse the first k nodes

3. Recursively call the function for the remaining list

4. Connect the reversed part with the rest

5. Return the new head


```python
# Definition for singly-linked list
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def reverse_k_group(head, k):
    # Step 1: Check if there are at least k nodes
    count = 0
    current = head
    while current and count < k:
        current = current.next
        count += 1

    # If fewer than k nodes, do not reverse
    if count < k:
        return head

    # Step 2: Reverse first k nodes
    prev = None
    curr = head
    for _ in range(k):
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node

    # Step 3: Recursively reverse remaining list
    head.next = reverse_k_group(curr, k)

    # prev is the new head after reversal
    return prev
```

##DaY 23

1. Create a dummy node
2. Set a pointer current to the dummy node

3. While both lists are not empty:
      - Compare list1.val and list2.val
      - Attach the smaller node to current
      - Move forward in the list you used
      - Move current forward

4. If one list is empty:
      - Attach the remaining nodes of the other list

5. Return dummy.next (the real head)


```python
# Definition for singly-linked list
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def merge_two_lists(list1, list2):
    # Dummy node to start the merged list
    dummy = ListNode()
    current = dummy

    # Compare nodes from both lists
    while list1 and list2:
        if list1.val <= list2.val:
            current.next = list1
            list1 = list1.next
        else:
            current.next = list2
            list2 = list2.next

        current = current.next

    # Attach remaining nodes
    if list1:
        current.next = list1
    else:
        current.next = list2

    # Return the merged list (skip dummy)
    return dummy.next
```

##Day 24

1. Create a dummy node
2. Set current pointer to dummy
3. Set carry = 0

4. While l1 exists OR l2 exists OR carry exists:
      - Get value from l1 (or 0 if none)
      - Get value from l2 (or 0 if none)
      - Add values + carry
      - New digit = sum % 10
      - New carry = sum // 10
      - Create a new node with new digit
      - Move pointers forward

5. Return dummy.next (head of result list)


```python
# Definition for singly-linked list
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def add_two_numbers(l1, l2):
    dummy = ListNode()
    current = dummy
    carry = 0

    # Loop while there are digits or a carry
    while l1 or l2 or carry:
        val1 = l1.val if l1 else 0
        val2 = l2.val if l2 else 0

        total = val1 + val2 + carry
        digit = total % 10
        carry = total // 10

        current.next = ListNode(digit)
        current = current.next

        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next

    return dummy.next
```

##Day 25

1. Create a dummy node that points to head
2. Set a pointer current to dummy

3. While there are at least two nodes ahead:
      - first = current.next
      - second = first.next

      - Swap the two nodes
      - Move current forward by two nodes

4. Return dummy.next (new head)


```python
# Definition for singly-linked list
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def swap_pairs(head):
    # Dummy node helps handle the head easily
    dummy = ListNode(0)
    dummy.next = head
    current = dummy

    # Loop while there are at least two nodes to swap
    while current.next and current.next.next:
        first = current.next
        second = first.next

        # Swapping
        first.next = second.next
        second.next = first
        current.next = second

        # Move to the next pair
        current = first

    return dummy.next
```
